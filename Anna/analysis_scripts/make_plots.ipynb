```python
# read in data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
apf_data = pd.read_csv('./SM_steller_properties/specmatch_results_all_apf.csv') # SP-EMP APF derived values
gaia_data =  pd.read_csv('./gaia_properties.csv') # gaia stellar properties (only R and Teff)
hires_data = pd.read_csv('/mnt_home/azuckerman/BL_APF_DAP/lincomb_results_detrend.csv') # SM-Emp values derived from HIRES spectra (detrended)
yee_data = pd.read_csv('/mnt_home/azuckerman/BL_APF_DAP/yee_library_full.csv') # library values

apf_Teff = apf_data['Teff_detrended'].to_numpy()
apf_R = apf_data['radius_detrended'].to_numpy()
apf_feh = apf_data['feh_detrended'].to_numpy()

gaia_Teff = gaia_data['teff_val'].to_numpy()
gaia_R = gaia_data['radius_val'].to_numpy()

yee_Teff = yee_data['Teff [K]'].to_numpy()
yee_R = yee_data['R* [Rsun]'].to_numpy()
yee_feh = yee_data['[Fe/H] [sun]'].to_numpy()
```

```python
# make plot after Fig. 1 of Yee et al 2017
plt.figure(figsize = [7,10])

plt.subplot(211)
plt.plot(apf_Teff, apf_R, '+', label = 'APF')
plt.plot(yee_Teff, yee_R, '+', label = 'Library')
plt.plot(gaia_Teff, gaia_R, '+', label = 'Gaia')
plt.xlabel('Effectice Temperature [K]')
plt.ylabel('Stellar Radius [Ro]')
plt.yscale('symlog')
plt.xlim([7000,3000])
plt.legend()

plt.subplot(212)
plt.plot(apf_feh, apf_R, '+', label = 'APF')
plt.plot(yee_feh, yee_R, '+', label = 'Library')
plt.xlabel('Stellar Metallicity [dex]')
plt.ylabel('Stellar Radius [Ro]')
plt.yscale('symlog')
plt.legend()

#plt.suptitle('APF SM-Emp Derived Values')
plt.tight_layout(pad=3.0)
```
