```python
# turn resutls csv file into a latex table snippet
# Run this cell to generate the latex code (it is the strnig output of this cell)
# either use that (copy past into a document), or run the last cell to convert to a csv 
# and upload the csv to https://www.tablesgenerator.com/latex_tables
# then change things (for this table I added a bar under the column names and centered everything.)

import pandas as pd
import numpy as np

results = pd.read_csv('/mnt_home/azuckerman/BL_APF_DAP/SM_stellar_properties/all_apf/specmatch_results_current.csv')
SM_results_df = pd.DataFrame(columns = ['$Name$', '$T_{eff} [K]$', '$R [R_\odot]$', '$[Fe/H] [dex]$', '$M [M_\odot]$', '$ log_{10}(age) $', '$log(g)$'], 
                   index = results.index)
iso_results_df = pd.DataFrame(columns = ['$Name$', '$T_{eff} [K]$', '$R [R_\odot]$', '$[Fe/H] [dex]$', '$M [M_\odot]$', '$log_{10}(age) $', '$log(g)$'], 
                   index = results.index)
j=0
for i in range(len(results)):
    
    row = results.iloc[i]
    if not (np.isnan(row['iso_Teff'])) and not row['iso_Teff'] == 0:
       # SM_results_df.iloc[i,:] =  [row['name'],'','','','','','']
       # iso_results_df.iloc[i,:] = [row['name'],'','','','','','']  
   # else:
        SM_results_df.iloc[j,:] =  [row['name'],
                                   '$' + str(int(np.round(row['Teff_detrended'],0))) + ' \pm ' + str(int(np.round(row['u_Teff_detrended'],0))) + '$',
                                   '$' +str('{:.2f}'.format(np.round(row['radius_detrended'],2))) + ' \pm ' + str('{:.3f}'.format(np.round(row['u_radius_detrended'],3))) + '$',    
                                   '$' +str('{:.2f}'.format(np.round(row['feh_detrended'],2))) + ' \pm ' + str('{:.3f}'.format(np.round(row['u_feh_detrended'],3)))+ '$', 
                                   '$' +str('{:.2f}'.format(np.round(row['mass_detrended'],2))) + ' \pm ' + str('{:.2f}'.format(np.round(row['u_mass_detrended'],2)))+ '$', 
                                   '$' +str('{:.2f}'.format(np.round(row['age_detrended'],2))) + ' \pm ' + str('{:.2f}'.format(np.round(row['u_age_detrended'],2)))+ '$', 
                                   '$' +str('{:.2f}'.format(np.round(row['logg_detrended'],2))) + ' \pm ' + str('{:.2f}'.format(np.round(row['u_logg_detrended'],2)))+ '$']
        iso_results_df.iloc[j,:] = [row['name'],                            
                                   '$' +str(int(np.round(row['iso_Teff'],0))) + '^{ + ' + str(int(np.round(row['iso_up_Teff'],0))) + ' } _ { - ' + str(int(np.round(row['iso_um_Teff'],0))) +'}$', 
                                   '$' +str('{:.2f}'.format(np.round(row['iso_radius'],2))) + '^{ + ' + str('{:.2f}'.format(np.round(row['iso_up_radius'],3))) + ' } _ { - ' + str('{:.2f}'.format(np.round(row['iso_um_radius'],3)))+'}$',  
                                   '$' +str('{:.2f}'.format(np.round(row['iso_feh'],2))) + '^{ + ' + str('{:.2f}'.format(np.round(row['iso_up_feh'],3))) + ' } _ { - ' + str('{:.2f}'.format(np.round(row['iso_um_feh'],3)))+'}$',
                                   '$' +str('{:.2f}'.format(np.round(row['iso_mass'],2))) + '^{ + ' + str('{:.2f}'.format(np.round(row['iso_up_mass'],2))) + ' } _ { - ' + str('{:.2f}'.format(np.round(row['iso_um_mass'],2)))+'}$', 
                                   '$' +str('{:.2f}'.format(np.round(row['iso_age'],2))) + '^{ + ' + str('{:.2f}'.format(np.round(row['iso_up_age'],2))) + ' } _ { - ' + str('{:.2f}'.format(np.round(row['iso_um_age'],2)))+'}$', 
                                   '$' +str('{:.2f}'.format(np.round(row['iso_logg'],2))) + '^{ + ' + str('{:.2f}'.format(np.round(row['iso_up_logg'],2))) + ' } _ { - ' + str('{:.2f}'.format(np.round(row['iso_um_logg'],2))) +'}$' ]    
        j+=1
#usecols = np.zeros(len(results.iloc[0,:]))
#usecols[0] = 1
#usecols[4] = 1
#usecols[14:24] = 1
results_abbreviated_SM = SM_results_df.iloc[250:300]#, usecols.astype(bool)] 
#results_abbreviated_SM .rename(columns={'': 'newName1', 'oldName2': 'newName2'}, inplace=True)
print(results_abbreviated_SM.to_latex(index = False))

results_abbreviated_iso = iso_results_df.iloc[250:300]#, usecols.astype(bool)] 
print(results_abbreviated_iso.to_latex(index = False))
```
