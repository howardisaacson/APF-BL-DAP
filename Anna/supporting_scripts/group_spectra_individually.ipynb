```python
# enclose each spectrum individually in a directory, in order to run on indiivual spectra rather than stacking.
# Simplest way to do this, becuase SM-Emp expects each target's spectra to be enclosed in a directory

import os
import shutil
import pandas as pd
import astropy.io.fits as fits 
```

```python
dirlist = os.listdir('APF_spectra/all_apf_spectra')

for star_dir in dirlist:
    if not star_dir.startswith('NGC'):
        star_name = star_dir.split('_')[0]
        idx = 0
        for file in os.listdir('APF_spectra/all_apf_spectra/' + star_dir):
            new_dir_name = star_name + '_spectra_' + str(idx)
            os.mkdir('APF_spectra/all_individual_spectra/' + new_dir_name)
            shutil.copy('APF_spectra/all_apf_spectra/' + star_dir + '/' + file, 'APF_spectra/all_individual_spectra/' + new_dir_name + '/' + file)
            idx += 1
```

```python
# Oops, actually shouldn't have inlcuded all the stars. Should exclude the ones with iodine, binaries, galaxies, etc. 
# 1) check if each sepctrum has iodine (bc flag in results file is for whether ANY of that star's spectra had iodine)
# 2) get list of stars that are binaries, or aren't stars, and list of spectra of those stars 
# Still will keep the direcoty with all the spectra, in case useful later, but will only run on these.

results = pd.read_csv('/mnt_home/azuckerman/BL_APF_DAP/SM_stellar_properties/all_apf/11Jan2022/specmatch_results_11Jan2022.csv')

for star_dir in os.listdir('APF_spectra/all_individual_spectra'):
    filename = os.listdir('APF_spectra/all_individual_spectra/' + star_dir)[0]
    star_name = filename.split('_')[0]
    file = fits.open('APF_spectra/all_individual_spectra/' + star_dir + '/' + filename)
    iodine = file[0].header['ICELNAM'] == 'In'
    non_star = results[results['Simbad_resolvable_name'] == star_name]['non-stellar_flag'].any()
    binary = results[results['Simbad_resolvable_name'] == star_name]['binary_flag'].any()
    if not (iodine | non_star | binary):
        os.mkdir('APF_spectra/all_individual_spectra_clean/' + star_dir)
        shutil.copy('APF_spectra/all_individual_spectra/' + star_dir + '/' + filename, 'APF_spectra/all_individual_spectra_clean/' + star_dir + '/' + filename)
```

```python
# just get the number w/ iodine

has_iodine = []
has_iodine_spect = []
for star_dir in os.listdir('APF_spectra/all_individual_spectra'):
    filename = os.listdir('APF_spectra/all_individual_spectra/' + star_dir)[0]
    star_name = filename.split('_')[0]
    file = fits.open('APF_spectra/all_individual_spectra/' + star_dir + '/' + filename)
    iodine = file[0].header['ICELNAM'] == 'In'
    if iodine:
        has_iodine += [star_name]
        has_iodine_spect += [filename]
        
print(len(set(has_iodine)))
print(len(has_iodine_spect))
```
