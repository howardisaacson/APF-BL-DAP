```python
# for APF log file
import pandas as pd
import numpy as np

def convert_type(type_key):
    if type_key == '*': long_type = 'Star'
    if type_key == 'PM*': long_type = 'High Proper-motion Star'
    if type_key == 'SB*': long_type = 'Spectroscopic Binary'
    if type_key == '**': long_type = 'Double or Multiple Star'
    if type_key == 'EB*': long_type = 'Eclipsing Binary'        
    if type_key == 'RS*': long_type = 'Variable Star' 
    if type_key == 'BY*': long_type = 'Variable Star' 
    if type_key == 'a2*': long_type = 'Variable Star' 
    if type_key == 'Er*': long_type = 'Eruptive variable Star'
    if type_key == 'dS*': long_type = 'Variable Star'
    if type_key == 'LP?': long_type = 'Long-period Variable Candidate'
    if type_key == 'Pu*': long_type = 'Pulsating variable Star'
    if type_key == 'GiG': long_type = 'Galaxy'
    if type_key == 'SyG': long_type = 'Galaxy'
    if type_key == 'PN': long_type = 'Planetary Nebula'
    if type_key == 'Em*': long_type = 'Emission-line Star'
    if type_key == 'RG*': long_type = 'Red Giant Branch Star'
    if type_key == 'El*': long_type = 'Variable Star'
    if type_key == 'Ro*': long_type = 'Variable Star'
    if type_key == 'gD*': long_type = 'Variable Star'      
    if type_key == 'Pe*': long_type = 'Peculair Star'    
    if type_key == 'HB*': long_type = 'Horizontal Branch Star'    
    if type_key == 'V*': long_type = 'Variable Star'
    if type_key == 'LM*': long_type = 'Low-mass Star'        
    if type_key == 'Not found': long_type = 'None'        

    return long_type

results = pd.read_csv('/mnt_home/azuckerman/BL_APF_DAP/apf_log_full_current.csv')
#log_results_df = pd.DataFrame(columns = ['Filename', 'Name','Date observed', 'Type', 'SNR', 'Filesize'], 
#                   index = results.index)
log_results_df = pd.DataFrame(columns = [ 'Name','Date observed', 'Type', 'SNR'], 
                   index = results.index)
for i in range(len(results)):
    row = results.iloc[i]
    obs_type = convert_type(row['Main_type'])
    log_results_df.iloc[i,:] = [row['Simbad_resolvable_name'],  
        row['Date_obs'],
        obs_type,
        int(row['SNR'])]#,
       # row['Filesize']]

results_abbreviated_log = log_results_df.iloc[250:300]#, usecols.astype(bool)] 
print(results_abbreviated_log.to_latex(index = False))
```
