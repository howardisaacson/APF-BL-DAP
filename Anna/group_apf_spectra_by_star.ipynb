```python
# Group all the BL APF spectra files into subdirectories by the star they belong to. 

import pandas as pd
import shutil
import os

apf_name_conversion = pd.read_csv('apf_name_conversion_updated.csv') # same as apf_name_conversion but has a header
apf_path = '/datag/blpd0/datax/apf'

#star_list = pd.read_csv('overlap_no_repeats.csv')
#star_list = star_list['Column2']   
star_list = list(dict.fromkeys(apf_name_conversion['HIP_NAME'])) # all unique HIP names of spectra in BL APF directory

def get_filenames(starname): 
    rows = apf_name_conversion[apf_name_conversion['HIP_NAME'] == starname] 
    filenames = rows["FILENAME"]
    return filenames

# find unique HIP names
#star_list_HIP = np.copy(star_list)
#for star in star_list:
#    star_rows = apf_name_conversion[apf_name_conversion['NAME'] == star]
#    star_list_HIP = star_list_HIP + star_rows['HIP_NAME'].tolist()
#star_list_HIP_unique = list(dict.fromkeys(star_list_HIP))
star_list_HIP_unique = star_list

not_found = []
raw_files = []
skipped = []
for star in star_list_HIP_unique: 
    filenames = get_filenames(star)
    #if len(filenames)>1:
    print(star+': ')
    print(filenames)
    #dir_path = apf_path + '/' + star + '_spectra'
    dir_path = './APF_spectra/all_apf_spectra/' + star + '_spectra'
    os.mkdir(dir_path)
    for file in filenames:
        if file.startswith('r'): 
            #shutil.move('/home/azuckerman/APF_spectra/apf_spectra/'+file+'.fits', dir_path)
            try:
                shutil.copy(apf_path + '/' + file + '.fits', dir_path)
            except FileNotFoundError:
                print(file + '.fits not found in BL APF directory.')
                not_found += [file]
        elif file.startswith('ucb'): 
            print('skipping raw file ' + file + '.fits')
            raw_files += [file]
        else:
            print('skipping file ' + file + ' for unkown reason.')
            skipped =+ [file]
```
