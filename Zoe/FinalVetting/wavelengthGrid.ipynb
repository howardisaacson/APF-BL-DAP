## Plot SM spectra in observed frame

```python
star = 'HIP14286'
```

```python
observed_wl = 5015
arr = np.round(wl_image, 1) == np.round(observed_wl, 1)
order = np.where(arr)[0][0]
apf_files = np.load(apf_directory + star + '.npy')
observed_files = np.load(observed_directory + star + '.npy')


file = observed_files[-1]
    
subfile = fits.open(file)
SM_flux = subfile[0].data
SM_wl = subfile[1].data

header = subfile[0].header
star = header['TOBJECT']
date = header['DATE-OBS']
print(file)
print(star)
print(date)
plt.plot(SM_wl, SM_flux, color='blue', label = 'SM Spectrum (HIP14286_3_NDRU.fits)')

file = apf_files[-1]

fl_file = fits.open(file)
obj = fl_file[0].header['TOBJECT']
date = fl_file[0].header['DATE-OBS']
print(obj)
print(date)
print(file)
fl_image = fl_file[0].data
flux = fl_image[order][:-1]
wl = wl_image[order]
plt.plot(wl, (flux / 20000), color='orange', label='APF Spectrum (rbhv.240.fits')
plt.axvline(5018.33, color='blue', ls='--', label='5018.33 A')
plt.axvline(5019.7, color='orange', ls='--', label='5019.7 A')

plt.title('Observed Frame SM Spectrum vs. APF Spectrum, Shift: ' + str(np.round(5019.7 - 5018.33, 2)))
plt.xlim(5010, 5020)
plt.legend()
plt.show()
```
