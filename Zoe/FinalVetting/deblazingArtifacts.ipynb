```python
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
import astropy.io.fits as fits
from glob import glob

plt.rcParams["figure.figsize"] = (20, 10)
plt.rcParams['font.size'] = '16'
```

```python
stars = ['HIP84862', 'HIP14286', 'HIP50384', 'HIP24332', 'HIP68030']
stars = ['HIP14286', 'HIP68030']
```

```python
star = 'HIP68030'
```

```python
blaze_funcs = pd.read_csv('/mnt_home/azuckerman/BL_APF_DAP/bstar_blaze_funcs.csv')
order31_blaze = np.array(list(blaze_funcs.iloc[1])[1:])
order31_blaze_wls = [5001.94, 5012., 5014.16, 5017.54, 5018.38, 5049.88, 5049.92]

order32_blaze = np.array(list(blaze_funcs.iloc[2])[1:])
order32_blaze_wls = [5049.88, 5049.92, 5083.26, 5107.36]
```

```python
res = pd.read_csv('/home/zoek/code/APF-BL-DAP/Zoe/FinalVetting/pers_res.csv')

apf_directory = '/home/zoek/code/APF-BL-DAP/Zoe/FinalVetting/apf_data_files/'

APF_wavelength_path = '/mnt_home/zoek/code/APF-BL-DAP/Zoe/APFTutorial/apf_wav.fits'
wl_file = fits.open(APF_wavelength_path)
wl_image = wl_file[0].data
```

```python
observed_wl = 5000
apf_files = np.load(apf_directory + star + '.npy')
arr = np.round(wl_image, 1) == np.round(observed_wl, 1)
order = np.where(arr)[0][0]

file = apf_files[1]
print(file)
print(order)

fl_file = fits.open(file)
obj = fl_file[0].header['TOBJECT']
date = fl_file[0].header['DATE-OBS']
fl_image = fl_file[0].data
flux = fl_image[order][:-1] 
flux = flux / np.percentile(flux, 99)

wl = wl_image[order]
plt.plot(wl, flux)

plt.plot(wl[3:-4], order31_blaze)

plt.title('APF Spectrum: Order ' + str(order))
plt.show()
```
