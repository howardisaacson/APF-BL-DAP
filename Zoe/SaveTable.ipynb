# Combine tables from running allstars
Iterate through all the individual folders corresponding to each spectrum, count the number of detections for that spectrum, and make one table containing all the spectra and number of detections.

```python
import glob
import pandas as pd
```

```python
# folder that contains all the subfolders of each spectra
# already ran on 1, 3, 4, 5, 6
folder_to_combine = '/mnt_home/zoek/code/APF-BL-DAP/Zoe/LaserDetectionResults2'
```

```python
folders = glob.glob(folder_to_combine + '/*_*')
```

```python
folders[901]
```

```python
df = pd.DataFrame(columns = ['Star', 'Index', '# Above Threshold', '# Gaussian Shaped', '# Wider than PSF'])
```

```python
for folder in folders:
    folder_name = folder
    folder_name = folder_name.replace(folder_to_combine + '/', '')
    folder_name = folder_name.replace('_', '')
    i = 0
    while folder_name[i].isdigit():
        i += 1
    index = folder_name[0:i]
    star = folder_name[i:]
    file = glob.glob(folder + '/*.csv')
    file = file[0]
    table = pd.read_csv(file)
    above_threshold = table['indicies'][0]
    gaussian_shaped = table['indicies'][1]
    wider_than_PSF = table['indicies'][2]
    if above_threshold == '[]':
        at_detections = 0
    else:
        at_detections = above_threshold.count(' ') + 1
    if gaussian_shaped == '[]':
        gs_detections = 0
    else:
        gs_detections = gaussian_shaped.count(' ') + 1
    if wider_than_PSF == '[]':
        wtp_detections = 0
    else:
        wtp_detections = wider_than_PSF.count(' ') + 1
    df = df.append({'Star' : star, 'Index' : index, '# Above Threshold' : at_detections, 
                    '# Gaussian Shaped' : gs_detections, '# Wider than PSF' : wtp_detections}, ignore_index=True)
    
```
