# Determine Gaussian Cutoff
#### Determine what chi-squared value to use for both the spectra and residuals when determining whether a candidate is Gaussian-like or not.

```python
import numpy as np
import matplotlib.pylab as plt

plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams.update({'font.size': 19})
```

```python
spectra_chi = np.load('spectra_chi_squared.npy')
spectra_widths = np.load('spectra_widths.npy')
spectra_chi = spectra_chi[spectra_chi >= 0]
log_spect = np.log(spectra_chi)

residuals_chi = np.load('residuals_chi_squared.npy')
residuals_widths = np.load('residuals_widths.npy')
residuals_chi = residuals_chi[residuals_chi >= 0]
log_resid = np.log(residuals_chi)

print(len(spectra_chi))
```

```python
percentile = 95
```

```python
plt.hist(spectra_chi[spectra_widths < 2.7], density=True, alpha=0.5, label='Wider than PSF')
plt.hist(spectra_chi[spectra_widths > 2.7], density=True, alpha=0.5, label='Narrower than PSF')
plt.legend()
plt.title('Chi-squared Vals for Spectra Wider/Narrower than PSF')
plt.xlabel('Chi-squared Value')
plt.ylabel('Percent Spectra')

print('Max chi-squared val for injections narrower than PSF: ' + str(max(spectra_chi[spectra_widths > 2.7])))
print('Max chi-squared val for injections wider than PSF: ' + str(max(spectra_chi[spectra_widths < 2.7])))
plt.show()
```

```python
# spect_cutoff = np.percentile(spectra_chi, percentile)

spect_cutoff = max(spectra_chi[spectra_widths < 2.7])
print('Chi-squared threshold: ' + str(spect_cutoff))

plt.hist(spectra_chi, density=True)
# plt.axvline(x = spect_cutoff, color = 'red', label = 'Chi-squared Threshold of ' + str(round(spect_cutoff, 1)))
plt.axvline(x = spect_cutoff, color = 'red', label = '95 Percentile')
# plt.title('Distribution of Chi-squared Values for Spectra')
plt.xlabel('Reduced Chi-squared Value')
plt.ylabel('Percent')
plt.title('Distribution of Chi-squared Vals for Spectra')
plt.legend()
plt.show()
```
